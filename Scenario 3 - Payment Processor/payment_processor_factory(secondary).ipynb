# Payment Processor SECONDARY
This will use the <font color="blue">Factory Method design pattern</font> for the purposes of showing an example of another design pattern and why it is inferior to Strategy.


I nearly chose this design pattern as the primary pattern, but after a lot of deliberation, I chose strategy as this design pattern, on object switch, changing payment method means creating a new object through the factory, while strategy allows it to be changed directly without assigning a differnet strategy, as factory focuses on object creation while strategy focuses on swapping behaviour.

Factory also has many more lines of code as it needs additional factory classes, increasing complexity which strategy simply does not need. All for the minor benefit of slightly lower coupling. Which is why Strategy meets the requiremnts better (though only just).

## Start by creating the Factor Method Interface

This is what all factory classes must implement (create_payment())

```python
# Factory Method interface
class PaymentFactory:
    """
    Factory Method base class.
    
    Defines the method that the factory creation subclasses must use
    This allows object creation in subclasses reducing coupling between client and payment classes
    """
    def create_payment(self):
        pass
```

## All payment methods

Each method has their own `pay()` methods. This is very similar to how the strategy example looks but without the parent class.

```python
# The payment method classes used by the factory. As shown, they do not inherit anything.
# While they can all inherit a pay class, this is not necessary for the factory method to work 
# in this example.
class CardPayment:
    """
    What will be created by the factory.
    Contains the payment processing logic.
    """
    def pay(self, amount):
        print("Processing card details...")
        print(f"Card payment: £{amount}")


class PayPalPayment:
    """
    What will be created by the factory.
    Contains the payment processing logic.
    """
    def pay(self, amount):
        print("Connecting to PayPal servers...")
        print(f"PayPal payment: £{amount}")


class CryptoPayment:
    """
    What will be created by the factory.
    Contains the payment processing logic.
    """
    def pay(self, amount):
        print("Initialising blockchain...")
        print("Routing through dark web...")
        print(f"Crypto payment: £{amount}")
```

## Factory builders

One factory class per method to demonstrate the design pattern better, and to show the reduced coupling.

This is where the additional complexity is added, making it slightly less elegant than the Strategy design pattern

```python
# Factory classes for each payment type. This can be done in a single factory class with a parameter,
# but this is easier to extend and understand for demonstration purposes
class CardPaymentFactory(PaymentFactory):
    """
    Creates CardPayment object.
    """
    def create_payment(self):
        return CardPayment()


class PayPalPaymentFactory(PaymentFactory):
    """
    Creates PayPalPayment object.
    """
    def create_payment(self):
        return PayPalPayment()


class CryptoPaymentFactory(PaymentFactory):
    """
    Creates CryptoPayment object.
    """
    def create_payment(self):
        return CryptoPayment()
```

## Client Context Class

This class is what the user interacts with when selecting a payment method, Basket delegates the creation to the correct factory and so does not need to know about the selected payment method. This reduces coupling by having the Basket depend on the factories rather then the payment method classes. Different payment methods can be selected at runtime, which will replace the one currently in the basket instance.

No payment functionality is here, it is instead delegated to the correct pre-built class.

This is again very similar to the strategy design pattern, and I think could be made to only create the factory when checkout() is called, but I am leaving it as is to demonstrate the differences between the two design patterns (even if it was that method, it would still show a difference)

```python
# Client class
class Basket:
    """
    The client using the Factory Method.

    The Basket does not need to know the exact payment class being created.
    It only interacts with the factory interface.
    """
    def __init__(self, amount):
        self.amount = amount
        self.payment = None # No selected method yet

    def select_payment_factory(self, factory):
        # The factory creates the payment object.
        # This is done per switch in payment method
        self.payment = factory.create_payment()

    def checkout(self):
        if not self.payment:
            print("Please select a payment method first!")
            raise ValueError("No payment method selected")

        self.payment.pay(self.amount)
```

## Example output prints and useage

Here I show some examples and their results

- A card payment of 92.50
- A paypal payment of 31.03 (after first being a crypto type)

```python
def factory_examples():
    # The factory creates the selected payment object.
    # The Basket does not depend on the original payment classes.
    print(" Card payment example:")
    basket = Basket(92.50)
    basket.select_payment_factory(CardPaymentFactory())
    basket.checkout()
    print("")

    # The payment type can be changed by creating a different object,
    # but this requires creating a new payment object through another factory.
    # This is why the Strategy pattern is more flexible, as it allows changing the 
    # strategy without creating a new object.
    print(" PayPal payment example, with switching:")
    basket = Basket(31.03)
    basket.select_payment_factory(CryptoPaymentFactory())
    basket.select_payment_factory(PayPalPaymentFactory())
    basket.checkout()


if __name__ == "__main__":
    factory_examples()
```
