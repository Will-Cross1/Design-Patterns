# Media Player

This will use the <font color="blue">Decorator design pattern</font> for the purposes of showing an example of another design pattern and why it is inferior to the Adapter design pattern for this scenario.

This pattern allows additional functionality to be attached to objects without modifying their class code. Unfortunately, the pattern does not solve the main concern of the media libraries all being third-party meaning they may have incompatible interfaces rendering this pattern useless. However, The Decorator pattern can nicely be used in conjuntion with the Adapter pattern to add decorators to the adapted classes allowing the converted interfaces to have additional functionality that perhaps the app creators want (like a watermark)

I decided to do a decorator example rather than strategy, because I have already got a strategy code example.

## Media libraries and interface

The MediaPlayer is used as a common interface for all concrete media players and decorators to use.
In a real application, the media players will be third party, but this design patter does not support that, so is ignored for this example (breaking the requirement)

```python
# Decorator design pattern

# Base media player interface
class MediaPlayer:
    def play(self):
        raise NotImplementedError

# Main media player for MP4
class MP4Player(MediaPlayer):
    def __init__(self, file_path):
        self.file_path = file_path

    def play(self):
        print(f"Playing video: {self.file_path}")
        
# Main media player for MP3
class MP3Player(MediaPlayer):
    def __init__(self, file_path):
        self.file_path = file_path

    def play(self):
        print(f"Playing audio: {self.file_path}")
```

## Decorator classes

the PlayerDecorator is the base parent class for the other decorators for them to inherit play

The additional decorator features both add differnt things to the decorator, and can be used in conjunction.

```python
# Concrete decorator
class PlayerDecorator(MediaPlayer):
    def __init__(self, wrapped_player):
        self.wrapped_player = wrapped_player

    def play(self):
        self.wrapped_player.play()


# subtitle decorator feature
class SubtitleDecorator(PlayerDecorator):
    def __init__(self, wrapped_player, subtitle_file):
        super().__init__(wrapped_player)
        self.subtitle_file = subtitle_file

    def play(self):
        super().play()
        print(f"Displaying subtitles from: {self.subtitle_file}")

# Dub decorator feature
class DubDecorator(PlayerDecorator):
    def __init__(self, wrapped_player, dub_language):
        super().__init__(wrapped_player)
        self.dub_language = dub_language

    def play(self):
        super().play()
        print(f"Applying dubbing: {self.dub_language}")
```

## Example output prints and useage

As shown below, the decorators can be used on anychild of the MediaPlayer because of the play() method, For some use cases this will not be useful (for example applying a water mark to a sound file), but other instances can be quite useful, such as aqpplying a dub to a video or a song.

Multiple decorators can be used at once as well. (swapping the order will only swap the execution order)

```python
# Example use
if __name__ == "__main__":
    # Basic video player
    video_player = MP4Player("film.mp4")
    video_player.play()
    print("")

    # Video player with subtitle decorator
    video_player_with_subtitles = SubtitleDecorator(
        MP4Player("film.mp4"),
        "film.srt"
    )
    video_player_with_subtitles.play()
    print("")

    # Video player with subtitle and dub decorators
    video_player_all = DubDecorator(SubtitleDecorator(
            MP4Player("film.mp4"),
            "film.srt"
        ),
        "English Dub"
    )
    video_player_all.play()
    print("")
    print("")
    
    
    # Basic music player
    music_player = MP3Player("song.mp3")
    music_player.play()
    print("")

    # Music player with dub decorator
    music_player_with_dub = DubDecorator(
        MP3Player("song.mp3"),
        "English Dub"
    )
    music_player_with_dub.play()
    print("")
```
